```python
import pandas as pd
```

```python
# Load data
df = pd.read_excel("Online_Retail.xlsx")
df
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Basic cleanup
df = df[df['CustomerID'].notnull()]  # Drop rows with no customer ID
df = df[df['Quantity'] > 0]          # Remove returns/refunds
df = df[df['UnitPrice'] > 0]         # Remove zero-priced items
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Drop duplicates if any
df.drop_duplicates(inplace=True)
```

```python
# Create a new TotalPrice column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
# Extract Year/Month for time-based analysis
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
```

```python
# Print df to ensure changes
df
```

```python
# Check data types
df.dtypes
```

```python
# Export cleaned data
df.to_csv("online_retail_cleaned.csv", index=False)
```
